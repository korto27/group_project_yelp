```python
group1 = new_df[new_df["price"] == 1]["rating"]
group2 = new_df[new_df["price"] == 2]["rating"]
group3 = new_df[new_df["price"] == 3]["rating"]
group4 = new_df[new_df["price"] == 4]["rating"]
```

```python
stats.f_oneway(group1, group2, group3, group4)
```
